```python
import math
import random
import pandas as pd
import numpy as np
from google.colab import files

# Subir el archivo de entrenamiento
uploaded_train = files.upload()

# Cargar el archivo de entrenamiento en un DataFrame de Pandas
file_name_train = list(uploaded_train.keys())[0]
train_data = pd.read_csv(file_name_train)

# Subir el archivo de test
uploaded_test = files.upload()

# Cargar el archivo de test en un DataFrame de Pandas
file_name_test = list(uploaded_test.keys())[0]
test_data = pd.read_csv(file_name_test)
```

```python
# Definir la función de predicción
def y_predict(a, b, c, x):
    return a * np.cos(b * x + c)

# Hiperparámetros
lr = 0.05
batch = 100
epochs = 3000

# Extraer los datos de entrenamiento
X_train = train_data['x'].values
Y_train = train_data['TARGET'].values

# Inicializar parámetros
a = random.random()
b = random.random()
c = random.random()
d = 0

# Gradiente descendente
for epoch in range(epochs):
    a_gradiente = 0
    b_gradiente = 0
    c_gradiente = 0
    e = 0

    for _ in range(batch):
        ix = int(random.uniform(0, len(X_train)))
        y_hat = y_predict(a, b, c, X_train[ix])
        error = Y_train[ix] - y_hat
        e += error ** 2

        a_gradiente += -2 * error * np.cos(b * X_train[ix] + c)
        b_gradiente += 2 * error * a * X_train[ix] * np.sin(b * X_train[ix] + c)
        c_gradiente += 2 * error * a * np.sin(b * X_train[ix] + c)

    a -= lr * a_gradiente / batch
    b -= lr * b_gradiente / batch
    c -= lr * c_gradiente / batch

    e = e / batch

# Resultados finales
print(f"Final parameters: a = {a}, b = {b}, c = {c}")

# Predecir en el conjunto de test
y_test_pred = y_predict(a, b, c, test_data['x'].values)

# Crear DataFrame con los identificadores de ejemplo y las predicciones
submission = pd.DataFrame({
    'example_id': test_data['example_id'],
    'TARGET': y_test_pred
})

# Guardar el archivo de envío en formato CSV
submission.to_csv('submission.csv', index=False)
print("Archivo de envío 'submission.csv' generado.")
files.download('submission.csv')
```
